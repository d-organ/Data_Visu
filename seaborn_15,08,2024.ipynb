```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
```

## Scatter Plot

```python
sns.scatterplot(x = "total_bill" , y = "tip" , data = tips)
plt.show()
```

```python
sns.regplot(data=tips , y="tip" , x="total_bill")
plt.show()
```

```python
#korelasyon ilişki, regresyon etki(nedenselliği açıklar)
```

```python
#çizgini etrafındaki gölge = güven aralığı
#gölge genişşe yanlış tahmin ihtimali artar
#gölge dar ise doğru tahmin ihtimali artar
```

```python
sns.scatterplot(data=tips , y="tip" , x="total_bill" , hue="sex")
plt.show()
```

```python
#hue=kategorik olmalı
```

```python
sns.scatterplot(data=tips , y="tip" , x="total_bill" , hue="smoker")
plt.show()
```

```python
sns.scatterplot(data=tips , y="tip" , x="total_bill" , hue="smoker" ,style="sex")
plt.show()
```

```python
sns.scatterplot(data=tips , y="tip" , x="total_bill" , hue="smoker" ,style="day")
plt.show()
```

```python
sns.scatterplot(x=tips["total_bill"] , y=tips["tip"]  , hue=tips.smoker ,style=tips.sex)
plt.show()
```

```python
fig , ax = plt.subplots(1,2, figsize =(10,5))
sns.scatterplot(x="total_bill" , y="tip" , data=tips , style="sex" , ax=ax[0])
sns.scatterplot(x="total_bill" , y="tip" , data=tips , hue="smoker" , ax=ax[1])
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10 ,5))
sns.scatterplot(x = "total_bill", y = "tip", data = tips, style = "sex", ax = ax[0])
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "smoker", ax = ax[1]);
```

```python
#palatte

plt.figure(figsize = (10,5))
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "sex", palette = "plasma")
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "sex", palette = "viridis")
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "sex", palette = "plasma" , alpha=0.5)
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "sex", palette = "plasma" , alpha=0.3)
plt.show()
```

```python
#yoğun noktayı daha rahat görebilmek için kullanıyoruz.
```

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(x = "total_bill", y = "tip", data = tips,
                hue = "sex",
                palette = "plasma",
                alpha = 0.7,
                s = 300)
plt.show()
```
